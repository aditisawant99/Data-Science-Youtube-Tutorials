```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

### Basic Graph
matplotlib documentation : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#Resize your graph
plt.figure(figsize=(9,7))

#line 1


#plot
#keyword argument notation
#plt.plot(x,y, label = "2x", color='magenta', linewidth=2, linestyle="--", marker="*", markersize=10, markeredgecolor="red")

#Use shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y, 'm*--', label = "2x") 

##Line 2

#select interval we want to place points at
x2 = np.arange(0,4.5,0.5) #from 0 to 4 with jump of 0.5, like sequence in R
print(x2)


#plt.plot(x2,x2**2, label='x^2', marker=".", color='orange')

#plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2, label='x^2',marker='.') #for plotting just the first six values

#plot remainder of the graph as dotted
plt.plot(x2[5:],x2[5:]**2, linestyle="--")

#titles and labels (using fontdict)
plt.title("Our first graph!", fontdict={'fontname': 'Comic Sans MS','fontsize':15})
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#remove decimal values (change ticks)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

#legend
plt.legend()

plt.savefig('mygraph1.png', dpi=200)
plt.show()
```

### Bar chart

```python
labels = ['A','B','C','D']
values = [1,2,4,9]
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('.')
bars[2].set_hatch('*')
bars[3].set_hatch('O')

patterns = ['/','.','*','O']

for bar in bars:
    bar.set_hatch(patterns.pop())
plt.show()
```
